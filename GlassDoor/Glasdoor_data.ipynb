```python
col = ['Atlanta' , 'Boston' , 'Chicago' , 'Houston' , 'Los Angeles' , 'New York City' , 'Philadelphia' , 'San Francisco' , 'Seattle','Washington DC']
df['Mean']= df.mean(axis =1)
col_me = ['Mean']
df_mean = df[col_me]
# df_mean = pd.DataFrame(df_mean, columns=['Mean'])
df_mean
```

```python
col = ['Job Category','Job Title','Atlanta' , 'Boston' , 'Chicago' , 'Houston' , 'Los Angeles' , 'New York City' , 'Philadelphia' , 'San Francisco' , 'Seattle','Washington DC']
df_std = df[col]
df_std['STD']= df_std.std(axis =1)
col_koop = ['STD']
df_std = df_std[col_koop]
df_std
```

```python
df_me_st = df_mean.join(df_std)
df_me_st['mean+std'] = df_me_st.sum(axis = 1)
df_me_st.head(10)
```

```python
df_me_st['Mean - STD'] = df_me_st['Mean']-df_me_st['STD']
```

```python
df_me_st
```

```python
col_jobcat= ['Job Category','Job Title']
df_jobcat = df[col_jobcat]
df_jobcat
```

```python
dfmeasnstdjob = df_jobcat.join(df_me_st)
dfmeasnstdjob
```

```python
df_mean_scat = px.scatter(dfmeasnstdjob, x="Job Category", y="STD", text ='Job Title', color='STD')
df_mean_scat.update_traces(textposition='top center')
df_mean_scat.update_layout(
    height=800,
    width =1000,
    title_text='Average salary for the Job Title in each Category'
)
df_mean_scat.show()
```
